```python
def adjacent_edges(nodes, halfk):
    n = len(nodes)
    for i, u in enumerate(nodes):
        for j in range(i+1, i+halfk+1):
            v = nodes[j % n]
            yield u, v
```

```python
def make_ring_lattice(n, k):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(adjacent_edges(nodes, k//2))
    return G
```

```python
lattice = make_ring_lattice(10, 4)
pos = nx.circular_layout(lattice)
plt.figure(figsize=(8,8))
nx.draw_networkx(lattice, pos=pos, with_labels = False)
plt.axis('off')
plt.title('Regular Ring Lattice Graph with N=10, K=4', fontsize = 20)
```

```python
def flip(p):
    return np.random.random() < p

def rewire(G, p):
    nodes = set(G)
    for u, v in G.edges():
        if flip(p):
            choices = nodes - {u} - set(G[u])
            new_v = np.random.choice(list(choices))
            G.remove_edge(u, v)
            G.add_edge(u, new_v)
    return G
```

```python
plt.figure(figsize=(36,10))

plt.subplot(131,)
G_l = rewire(lattice, p=0)
pos_l = nx.circular_layout(G_l)
nx.draw(G_l,pos_l, with_labels = False)
plt.title('Rewiring with Beta=0', fontsize = 30)

plt.subplot(132)
G_ws = rewire(lattice, p=0.3)
pos_ws = nx.circular_layout(G_ws)
nx.draw(G_ws,pos_ws, with_labels=False)
plt.title('Rewiring with Beta=0.3', fontsize = 30)


plt.subplot(133)
G_r = rewire(lattice, p=1)
pos_r = nx.circular_layout(G_r)
nx.draw(G_r,pos_r, with_labels=False)
plt.title('Rewiring with Beta=1', fontsize = 30)



```
