# Three functions without marker

```python
def symmetric_log_loss(y, p):
    return -(y*np.log(p) + (1.-y)*np.log(1.-p))

def rise_risk_log_loss(y, p, b=2):
    return -(b*y*np.log(p) + (1.-y)*np.log(1.-p))

def reduce_risk_log_loss(y, p, g=2):
    return -(y*np.log(p) + g*(1.-y)*np.log(1.-p))
```

```python
import matplotlib.pyplot as plt
import numpy as np

f = plt.figure(figsize=(20,5))
ax = f.add_subplot(131)
ax2 = f.add_subplot(132)
ax3 = f.add_subplot(133)



# 100 linearly spaced numbers
p = np.linspace(0,1,100)

# the function, which is y = x^2 here
y = symmetric_log_loss(1, p)
y2 = symmetric_log_loss(0, p)



# plot the function
ax.plot(p,y, 'r', label='y=1')
ax.plot(p, y2, 'b', label = 'y=0')
ax.set_ylim([0,7])
ax.legend()
ax.set_title('Symmetric log-loss')

y = rise_risk_log_loss(1, p)
y2 = rise_risk_log_loss(0, p)

ax2.plot(p,y, 'r', label='y=1')
ax2.plot(p, y2, 'b', label = 'y=0')
ax2.set_ylim([0,7])
ax2.legend()
ax2.set_title('Asymmetric log-loss: biased towards high risk perception')

y = reduce_risk_log_loss(1, p)
y2 = reduce_risk_log_loss(0, p)


# plot the function
ax3.plot(p,y, 'r', label='y=1')
ax3.plot(p, y2, 'b', label = 'y=0')
ax3.set_ylim([0,7])
ax3.legend()
ax3.set_title('Asymmetric log-loss: biased towards low perception')




# show the plot
plt.show()
```
